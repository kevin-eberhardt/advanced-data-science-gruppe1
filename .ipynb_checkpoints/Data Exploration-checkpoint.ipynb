# Data Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data/BikeRentalDaily_train_valid.csv', sep = ";")
print("CSV-Datei:")
data.head(20)
```

## Erste Einblicke in das Dataset
Zunächst wollen wir uns einen Überblick über die Daten beschaffen. Da pandas hier für jedes DataFrame die Methode describe() bietet, können wir mithilfe dieser Funktion uns anschauen, wie die Daten beschaffen sind.


Aus den Vorlesungsdokumenten geht hervor, dass es sich bei dem Dataset um ein Dataset handelt, dass sich um die Nutzung von Bike-Sharing Angeboten handelt.
Übersicht über die Daten, die darin vorkommen:

| Attribut        | Beschreibung                                                                                                                                                                                                                                                          |
|-----------------|-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| instant         | Eintragsindex                                                                                                                                                                                                                                                         |
| dteday          | Datum                                                                                                                                                                                                                                                                 |
| season          | Saison (Winter = 1, Frühling = 2, Sommer = 3, Herbst = 4)                                                                                                                                                                                                             |
| yr              | Jahr                                                                                                                                                                                                                                                                  |
| mnth            | Monat (0 bis 12)                                                                                                                                                                                                                                                      |
| holiday         | Ob der Tag ein Feiertag ist(1) oder nicht (0)                                                                                                                                                                                                                         |
| weekday         | Wochentag                                                                                                                                                                                                                                                             |
| workingday      | Ob der Tag ein Arbeitstag ist (1) oder ein Wochenendstag bzw. Feiertag (0)                                                                                                                                                                                            |
| weathersit      | Wetterlage: 1: Klar, teilweise bewölkt, 2: Neblig, teilweise bewölkt, 3: Leichter Schnee oder leichter Regen oder Gewitter & verstreute Wolken oder leichter Regen und verstreute Wolken 4: Starker Regen und Eisplatten und Gewitter und Nebel oder Schnee und Nebel |
| temp            | Temperatur                                                                                                                                                                                                                                                            |
| atemp           | Gefühlte Temperatur                                                                                                                                                                                                                                                   |
| hum             | Luftfeuchtigkeit                                                                                                                                                                                                                                                      |
| windspeed       | Windgeschwindigkeit                                                                                                                                                                                                                                                   |
| leaflets        | Anzahl an Broschüren (für Marketingzwecke)                                                                                                                                                                                                                            |
| price reduction | Ob an diesem Tag Preis-Reduzierungen vorgenommen wurden (Ja: 1, Nein: 0)                                                                                                                                                                                              |
| casual          | Anzahl an zufälligen/neuen Benutzern                                                                                                                                                                                                                                  |
| registered      | Anzahl an registrierten Benutzern                                                                                                                                                                                                                                     |
| cnt             | Anzahl an vermieteten Fahrrädern (sowohl von zufällig als auch von registrierten Benutzern)                                                                                                                                                                           |

```python
import matplotlib.pyplot as plt
import seaborn as sns

g = sns.FacetGrid(data, row="weathersit", col="season", margin_titles=True)
g.map(sns.regplot, "temp", "cnt", color=".3", fit_reg=False, x_jitter=.1);
```
