```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)


N = 50 #N=50 data points
x = np.random.rand(N) #x position of 50 data points: rand(N) created N value in range[0;1]
y = np.random.rand(N) # y position of 50 data points
colors = np.random.rand(N) #colors of 50 data points
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

#alpha: blending value, between 0 (transparent) and 1 (opaque)
#s: marker size in points**2
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
#change marker shape and transparency
plt.scatter(x, y, s=area, c=colors, marker='s',alpha=0.2)
plt.show()
```

```python
#Additional features added
plt.scatter(x, y, s=area, c=colors, marker='v',label=colors, alpha=0.2)
plt.xlabel('x-feature')
plt.ylabel('y-feature')
plt.title('Scatter Title')
plt.grid(True)
plt.colorbar()
plt.show()
```

```python
# Most simple histogram example with the values of x
plt.hist(x)
plt.show()
```

```python
#Histogram example using some of the optional parameters
#weights:An array of weights, of the same shape as x. Each value in x only contributes its associated weight 
#towards the bin count (instead of 1), so with N=50, each value contributes 0.02 resulting in relative values
#in the histogram
plt.hist(x,bins=20,weights=np.zeros_like(x) + 1. / x.size,color='g')
plt.xlabel('Binned x values')
plt.ylabel('Fraction of value')
plt.title('My Green Histogram')
plt.show()
```

```python
# Most simple boxplot example with the values of x
plt.boxplot(x)
plt.show()
```

```python
#Boxplot example using some of the optional parameters
#Adding some outliers to the data
def neg_or_pos(x): #Function will be uses to create positive AND negative outliers in specific range
    coin=np.random.choice((-1, 1))
    if coin == -1:
        return(x*-1+1)
    else:
        return x

# Fixing random state for reproducibility
np.random.seed(19680801)

N = 50 #N=50 data points
x = np.random.rand(N) #x position of 50 data points: rand(N) created N value in range[0;1]
y = np.random.rand(N) # y position of 50 data points
    
#Create some outliers for x
N_outliers=10
v_neg_or_pos = np.vectorize(neg_or_pos)
x_out = np.random.rand(N_outliers)*3+2 #creates an array of N_outliers in range [2;5]
x_out = v_neg_or_pos(x_out)
#print(x_out)
x= np.append(x,x_out)

data = [x,y*2]
plt.boxplot(data,notch=True)
plt.xlabel('The two features')
plt.ylabel('Value range')
plt.title('Histogram with two features - one with outliers, one w/o')
plt.show()
```
