## Erste Einblicke in das Dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

g = sns.FacetGrid(data, row="weathersit", col="season", margin_titles=True)
g.map(sns.regplot, "temp", "cnt", color=".3", fit_reg=False, x_jitter=.1);
```

```python
sns.set_theme()
plt.stackplot(data["dteday"],data["cnt"])
plt.legend(loc='upper left')
plt.show()
```

Erkenntnis: Es gibt in der cnt Datensätze, die sehr stark ausreisen. Auch ersichtlich in der ersten Visualisierung: Manche Werte sind zwischen 30.000 und 50.000

Idee: Sichten dieser möglichen Aureiser-Werte.

```python
cnt_outliers = data[(data.cnt >= 30000)]
cnt_outliers.head(10)
```

Fragestellung: Warum ist "cnt" für diese Werte so hoch?

Genauere Untersuchung für diese Ausreiser-Werte:

```python
weekdays = ["Montag", "Dienstag", "Mittwoch", "Donnerstag", "Freitag", "Samstag", "Sonntag"]
weather_conditions = ["n.a.", "Klar, teilweise bewölkt", "Neblig, teilweise bewölkt", "Leichter Schnee oder leichter Regen oder Gewitter & verstreute Wolken oder leichter Regen und verstreute Wolken", "Starker Regen und Eisplatten und Gewitter und Nebel oder Schnee und Nebel"]
for i in range(0, len(cnt_outliers)):
    max_value = cnt_outliers.iloc[i]
    print("Ausreiser-Wert #", max_value.instant, " war am ", max_value.dteday, "und war ein", weekdays[max_value.weekday])
    print("\tWetter-Verhältnisse:\t\t", round(max_value.temp,2), "°C, gefühlt wie ", str(round(float(max_value.temp)*float(max_value.atemp), 2)), "°C")
    print("\t\tWetter-Lage:\t\t", weather_conditions[max_value.weathersit])
    print("\t\tPreis-Reduzierung?\t", "Ja" if max_value["price reduction"] == 1 else "Nein")
    print("\t\tZufällige Besucher:\t", max_value.casual, "(", str(round(float(max_value.casual)/float(max_value.cnt), 2)*100), "%)")
    print("\t\tRegistrierte Besucher:\t", max_value.registered, "(", str(round(float(max_value.registered)/float(max_value.cnt), 2)*100), "%)")
    print("\t\tBroschüren:\t\t", max_value.leaflets) 
```
